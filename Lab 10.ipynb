```python
#for 4046
hc = 12400 #eV/A
l = 4046.6 #A
lmax = 8092.44/2 #A
lmin = 8090.96/2 #A
deltal = (lmax-lmin)/2
#print(deltal)
deltaE = (hc*deltal)/(l**2)
#print(deltaE)

mj = 1
g = 2
muB = 5.7884e-5 #ev/T
B0 = deltaE/(mj*g*muB)
print("B0 = ",B0)
```

```python
#for 4358
l2 = 4358.4

lmaxin = 8715.57/2
lminin = 8715.13/2
deltalin = (lmaxin-lminin)/2
deltaein = (hc*deltalin)/(l2**2)

lmaxmid = 8716.01/2
lminmid = 8714.74/2
deltalmid = (lmaxmid-lminmid)/2
deltaemid = (hc*deltalmid)/(l2**2)

lmaxout = 8716.22/2
lminout = 8714.51/2
deltalout = (lmaxout-lminout)/2
deltaeout = (hc*deltalout)/(l2**2)
print("Changes in Energy (in, mid, out) = ")
print(deltaein)
print(deltaemid)
print(deltaeout)

gIn = deltaein/(mj*muB*B0)
gMid = deltaemid/(mj*muB*B0)
gOut = deltaeout/(mj*muB*B0)
print("g values (in 1/2, mid 3/2, out 5/2) = ")
print(gIn)
print(gMid)
print(gOut)
```

```python
sigdeltaL = 0.01
sigdeltaE = hc*sigdeltaL/(l**2)
sigB0 = sigdeltaE/(mj*g*muB)
print("uncertainty in magnetic field = ",sigB0)
```

4. According to the selection rules, the change in the magnetic quantum number is at most +/-1. At a wavelength of 4358.4 A, we won't see the lines from -1 to +1, +1 to -1, and we also won't see the line from 0 to 0 since it has a short lifetime and there is a rule that you won't see the 0 to 0 line. For the same wavelength, we saw the lines from -1 to 0, -1 to -1, 0 to 1, 0 to -1, 1 to 1, and 1 to 0 because they fall within the bounds specified by the selection rules. For 4046.6 A, we saw all of the transitions because they were all within the bounds specified by the selection rules.
